## Pre-processing of data

<p> As we can see that there are null values and and the adress is in string/object form.
    
    1. We will drop all the null values ( rows)
    2. Convert adress to longitute and latitude
    3. Drop date column as it does not have any effect on the price
</p>

```python
dataset =  pd.read_csv("estate_data1.csv")
```

```python
dataset.drop("date", axis=1, inplace=True)
dataset.drop("address", axis=1, inplace =True)
dataset.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
dataset.head()
```

```python

# importing the module
import numpy as np


# function to add latitude
def latitude(row):
    if row == "(nan, nan)":
        return np.nan
    else:
        return float(row[1:-1].split(",")[0])

# function to add longitude
def longitude(row):
    if row == "(nan, nan)":
        return np.nan
    else:
        return float(row[1:-1].split(",")[1])

    
# calling function
dataset["latitude"] = dataset["latitude, longitude"].apply(latitude)
dataset["longitude"] = dataset["latitude, longitude"].apply(longitude)

dataset.head()
```

```python
dataset.drop('latitude, longitude', axis=1, inplace=True)
dataset.head(3)
```

```python
temp = dataset.price
dataset.drop("price", axis=1, inplace =True)
dataset['price'] = temp
```
